## **Compute CDF, Plot CDF**

```
cdf_age = Cdf(df['age'])  #using the CDF function from Empiricaldist module
```

```
cdf_age.plot()
plt.title('CDF of age PLOT')
```

```
cdf_status.plot()
plt.title('CDF of Status plot')
```

## **Normalize the CDF**

```
cdf_age.normalize()
```

```
cdf_status.normalize()
```

```
cdf_age.step()  # -- CDF plot with step function
```

```
cdf_status.step()
```

## **Pair Plot**

```
age = sns.pairplot(df,hue='age')
```

```
age_less_45 = sns.pairplot(df.query('age < 45'),hue='age')  # with queries and conditions on columns
```

# **Box Plot**

```
sns.boxplot(x=df['status'], y=df['age'], data=df)
```


# **PairGrid - New to explore**



```
pairGrid = sns.PairGrid(df, vars=['age','year','status','nodes'])
pairGrid = pairGrid.map_diag(plt.hist)
pairGrid = pairGrid.map_offdiag(plt.scatter)
```

# **Facet Grid**

```
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="status", size=4) \
   .map(plt.scatter, "year", "nodes") \
   .add_legend();
plt.show();
```
